```python
import matplotlib.pyplot as plt
import numpy as np
```

# Step function

```python
from pylab import *
t=linspace(-10,10,1000)
N=len(t)
A=1
shift=0
u=zeros(N)
for i in range(0,N):
    if t[i]>=-shift:
        u[i]=A
plt.plot(t,u)
```

# Sigmoid Function

```python
f = 1 / (1 + np.exp(-x))
plot_graph(f, "Sigmoid(X)")
```

# Tanh Function

```python
f = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(f, "TanH(X)")
```

# ReLU Function

```python
def relu(x):
    if x >= 0:
        return x
    else:
        return 0
f = list(map(relu, x))
plot_graph(f,"ReLU(X)")
```

# ELU Function

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
f = elu(x)
plot_graph(f,"ELU(X)")
```

# SELU Function

```python
def selu(x, alpha=1.67326, scale=1.0507):
    mask = (x > 0).astype(float)
    return scale * (mask * x + alpha * (1 - mask) * (np.exp(x) - 1))
f = selu(x)
plot_graph(f,"SELU(X)")
```
